```python
import pandas as pd
import requests
import lxml.html as lh
```

```python
url='http://www.watertribe.com/Events/ChallengeResults.aspx'
```

```python
# get all the tables on the page

read_table = pd.read_html(url)
```

```python
# working first the default form selection...
# slice out the results from the list of dataframes returned by read_table and fix header

results = read_table[5][1:] 
results.columns = read_table[5].iloc[0]
```

```python
results #how's it look?
```

```python
# next, see if we can capture all the events that are in the dropdown to later drive a selection

#Create a handle, page, to handle the contents of the website
page = requests.get(url)

#Store the contents of the website under doc
doc = lh.fromstring(page.content)

#Parse data that are stored @ the xpath of the dropdown
dropdown = doc.xpath('//*[@id="ctl00_ContentPlaceHolder1_DropDownListSelChallenges"]')
```

```python
# create a list of all the events available in the form dropdown
events = []

for i in dropdown[0]: 
    events.append(i.text_content())
```
